```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 가정: 'df'는 시계열 데이터가 인덱스이고 다른 여러 컬럼들이 있는 데이터 프레임입니다.

# 데이터를 numpy array로 변환
scaler2 = StandardScaler()
x = scaler2.fit_transform(data1.values)
# x = data1.values
# X = data.values

# 클러스터의 개수를 1부터 10까지 변화시키면서 KMeans 클러스터링 수행
inertias2 = []
for k in range(1, 11):
    kmeans2 = KMeans(n_clusters=k, random_state=0).fit(x)
    inertias2.append(kmeans2.inertia_)

# 클러스터의 개수에 따른 Inertia 값 그래프 그리기
plt.plot(range(1, 11), inertias2, 'o-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# 최적의 클러스터 개수 선택 (예: 3)
optimal_k = 4

# 최적의 클러스터 개수로 K-means 클러스터링 수행
kmeans2 = KMeans(n_clusters=optimal_k, random_state=0).fit(x)

# 클러스터링 결과 가져오기
labels2 = kmeans2.labels_

from collections import Counter

# 레이블 별 개수 확인
label_counts2 = Counter(labels2)

# 레이블 별 개수 출력
for label, count in label_counts2.items():
    print(f'Label {label}: {count}개')
```
