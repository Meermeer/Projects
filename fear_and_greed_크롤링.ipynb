```python
# 필요한 라이브러리들을 로드하자
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
import undetected_chromedriver as uc
```

```python
# 인베스팅닷컴 열기
driver = uc.Chrome()
url = 'https://pyinvesting.com/fear-and-greed/'  # 원하는 웹사이트 주소로 변경하세요
driver.get(url)

actions = ActionChains(driver)
```

```python
element2 = driver.find_elements(By.XPATH, '//*[@transform="translate(10,0)"]')
actions.click(element2[0]).perform()
```

```python
# 첫 데이터가 나타나는 곳으로 커서를 이동후 클릭(내가 수동으로 찾음)
actions.move_by_offset(-16, 180).perform()
```

```python
actions.move_by_offset(400, 0).perform()
```

```python
data_list = []

for i in range(200):

    html = driver.page_source
    # BeautifulSoup 객체를 생성
    soup = bs(html, 'html.parser')
    # 클래스가 "chart-wrapper"인 요소를 찾아서 해당하는 것을 data에 저장
    wrapper = soup.find('g', attrs={'class': 'highcharts-label highcharts-tooltip-box highcharts-color-undefined'})
    wrapper2 = soup.find('g', attrs={'class': 'highcharts-label highcharts-tooltip-header highcharts-tooltip-box'})
    data = wrapper.prettify()
    data2 = wrapper2.prettify()

    number_start_index = data.find('<tspan style="font-weight: bold;">\n') + len('<tspan style="font-weight: bold;">\n')
    number_end_index = data.find('<tspan class="highcharts-br">\n')
    number = data[number_start_index:number_end_index].strip()
    number = number.replace('\n  </tspan>','')

    date_start_index = data2.find('<tspan style="font-size: 0.8em;">\n') + len('<tspan style="font-size: 0.8em;">\n')
    date_end_index = data2.find("</tspan>", date_start_index)
    date = data2[date_start_index:date_end_index].strip()

    # 추출한 정보를 리스트에 형태로 하나씩 추가
    data_list.append({'date': date, 'number': number})

    # 커서를 한 칸씩 옮겨보자
    actions.move_by_offset(1, 0).click().perform()

    print(f'{i}번째 읽어옴')

# 커서를 하나씩 이동 시켜서 하나씩 크롤링 하는 작업을 끝냈으면 해당 데이터를 데이터 프레임으로 저장
df = pd.DataFrame(data_list)
```

```python
# 생성한 데이터 프레임에서 중복되는 값은 삭제
df_droped = df.drop_duplicates(subset='date')

# 그리고 중복되는 데이터를 제거한 완성된 데이터를 csv의 파일형태로 저장해둔다
df_droped.to_csv('fear_and_greed.csv', index=False)
```
